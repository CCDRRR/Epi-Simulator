设置一个虚拟的N×N的表格当作城市。这个城市里有不同的场所，分为home, workplace这两种。
agent 和 place 分别为两个class 没有learning和memory
只考虑home和workplace,默认感染率相同。每个人有一个workplace ID 和home ID, 白天的时候他们出去workplace上班，晚上回家。先假设每个人的感染概率一样都是60%，一开始SEIRD中只有一个人感染。对于在workplace和家里，都采用抽样的方法，从0-100中选个随机数，如果小于感染概率这个人就会感染，如果大于感染概率这个人就不会感染。可以1-6人组成家庭，住在一个home里。

```python
import numpy as np
import random
from mesa import Agent, Model
from mesa.time import RandomActivation
from mesa.space import MultiGrid
import matplotlib.pyplot as plt
```

```python
class Place:
    def __init__(self, place_id, place_type):
        self.place_id = place_id
        self.place_type = place_type
        self.agents = []

    def add_agent(self, agent):
        self.agents.append(agent)

    def remove_agent(self, agent):
        self.agents.remove(agent)

    def simulate_infection(self, infection_probability):
        for agent in self.agents:
            if agent.health_status == "Susceptible":
                if random.uniform(0, 100) < infection_probability:
                    agent.health_status = "Exposed"
```

```python
class CityAgent(Agent):
    def __init__(self, unique_id, model, home_id, work_id):
        super().__init__(unique_id, model)
        self.home_id = home_id
        self.work_id = work_id
        self.health_status = "Susceptible"
        self.days_infected = 0

    def step(self):
        if self.model.time_of_day == "morning":
            self.model.places[self.home_id].remove_agent(self)
            self.model.places[self.work_id].add_agent(self)
            self.model.grid.move_agent(self, self.model.place_positions[self.work_id])
        elif self.model.time_of_day == "evening":
            self.model.places[self.work_id].remove_agent(self)
            self.model.places[self.home_id].add_agent(self)
            self.model.grid.move_agent(self, self.model.place_positions[self.home_id])
        
        self.update_health_status()

    def update_health_status(self):
        if self.health_status == "Exposed":
            self.days_infected += 1
            if self.days_infected >= self.model.latency_period:
                self.health_status = "Infected"
                self.days_infected = 0
        elif self.health_status == "Infected":
            self.days_infected += 1
            if self.days_infected >= self.model.infection_duration:
                if random.random() < self.model.recovery_rate:
                    self.health_status = "Recovered"
                else:
                    self.health_status = "Deceased"
```

```python
class CityModel(Model):
    def __init__(self, N, num_agents, infection_probability, latency_period, infection_duration, recovery_rate):
        self.num_agents = num_agents
        self.grid = MultiGrid(N, N, True)
        self.schedule = RandomActivation(self)
        self.time_of_day = "morning"
        self.infection_probability = infection_probability
        self.latency_period = latency_period
        self.infection_duration = infection_duration
        self.recovery_rate = recovery_rate

        self.places = {}
        self.place_positions = {}

        # Create homes and workplaces
        for i in range(N):
            for j in range(N):
                place_id = i * N + j
                if random.random() < 0.5:
                    place_type = "home"
                else:
                    place_type = "workplace"
                place = Place(place_id, place_type)
                self.places[place_id] = place
                self.place_positions[place_id] = (i, j)

        # Create agents
        for i in range(self.num_agents):
            home_id = random.choice(list(self.places.keys()))
            work_id = random.choice(list(self.places.keys()))
            agent = CityAgent(i, self, home_id, work_id)
            self.schedule.add(agent)
            self.grid.place_agent(agent, self.place_positions[home_id])
            self.places[home_id].add_agent(agent)

        # Infect the first agent
        first_agent = self.schedule.agents[0]
        first_agent.health_status = "Infected"

    def step(self):
        self.schedule.step()

        # Simulate infections in all places
        for place in self.places.values():
            place.simulate_infection(self.infection_probability)

        # Toggle time of day
        if self.time_of_day == "morning":
            self.time_of_day = "evening"
        else:
            self.time_of_day = "morning"
```

```python
N = 10  # 10x10 grid
num_agents = 100
infection_probability = 60  # 60%
latency_period = 3  # 3 days of latency period
infection_duration = 14  # 14 days of infection duration
recovery_rate = 0.95  # 95% recovery rate

model = CityModel(N, num_agents, infection_probability, latency_period, infection_duration, recovery_rate)

for i in range(20):  # Simulate 20 steps
    model.step()
    print(f"Step {i+1}")
    for agent in model.schedule.agents:
        print(f"Agent {agent.unique_id} is {agent.health_status}")

# 可视化最终的感染状态
def plot_infection(model):
    grid = np.zeros((N, N))
    for agent in model.schedule.agents:
        x, y = agent.pos
        if agent.health_status == "Infected":
            grid[x][y] = 1
        elif agent.health_status == "Exposed":
            grid[x][y] = 0.75
        elif agent.health_status == "Recovered":
            grid[x][y] = 0.5
        elif agent.health_status == "Deceased":
            grid[x][y] = 0.25
    plt.imshow(grid, cmap="viridis")
    plt.title("Infection Status (Yellow = Infected, Light Green = Exposed, Dark Green = Recovered, Brown = Deceased)")
    plt.show()

plot_infection(model)
```

```python
import numpy as np
import random
from mesa import Agent, Model
from mesa.time import RandomActivation
from mesa.space import MultiGrid
import matplotlib.pyplot as plt
```

```python
class Place:
    def __init__(self, place_id, place_type):
        self.place_id = place_id
        self.place_type = place_type
        self.agents = []

    def add_agent(self, agent):
        self.agents.append(agent)

    def remove_agent(self, agent):
        self.agents.remove(agent)

    def simulate_infection(self, infection_probability, mask_effectiveness):
        for agent in self.agents:
            if agent.health_status == "Susceptible":
                effective_infection_probability = infection_probability
                if agent.wearing_mask:
                    effective_infection_probability *= (1 - mask_effectiveness)
                if random.uniform(0, 100) < effective_infection_probability:
                    agent.health_status = "Exposed"
```

```python
class CityAgent(Agent):
    def __init__(self, unique_id, model, home_id, work_id):
        super().__init__(unique_id, model)
        self.home_id = home_id
        self.work_id = work_id
        self.health_status = "Susceptible"
        self.days_infected = 0
        self.wearing_mask = False

    def step(self):
        if self.model.time_of_day == "morning":
            self.model.places[self.home_id].remove_agent(self)
            self.model.places[self.work_id].add_agent(self)
            self.model.grid.move_agent(self, self.model.place_positions[self.work_id])
            self.wearing_mask = True  # 必须在工作地点戴口罩
        elif self.model.time_of_day == "evening":
            self.model.places[self.work_id].remove_agent(self)
            self.model.places[self.home_id].add_agent(self)
            self.model.grid.move_agent(self, self.model.place_positions[self.home_id])
            self.wearing_mask = False  # 在家不戴口罩

        self.update_health_status()

    def update_health_status(self):
        if self.health_status == "Exposed":
            self.days_infected += 1
            if self.days_infected >= self.model.latency_period:
                self.health_status = "Infected"
                self.days_infected = 0
        elif self.health_status == "Infected":
            self.days_infected += 1
            if self.days_infected >= self.model.infection_duration:
                if random.random() < self.model.recovery_rate:
                    self.health_status = "Recovered"
                else:
                    self.health_status = "Deceased"
```

```python
class CityModel(Model):
    def __init__(self, N, num_agents, infection_probability, latency_period, infection_duration, recovery_rate, mask_effectiveness):
        self.num_agents = num_agents
        self.grid = MultiGrid(N, N, True)
        self.schedule = RandomActivation(self)
        self.time_of_day = "morning"
        self.infection_probability = infection_probability
        self.latency_period = latency_period
        self.infection_duration = infection_duration
        self.recovery_rate = recovery_rate
        self.mask_effectiveness = mask_effectiveness

        self.places = {}
        self.place_positions = {}

        # Create homes and workplaces
        for i in range(N):
            for j in range(N):
                place_id = i * N + j
                if random.random() < 0.5:
                    place_type = "home"
                else:
                    place_type = "workplace"
                place = Place(place_id, place_type)
                self.places[place_id] = place
                self.place_positions[place_id] = (i, j)

        # Create agents
        for i in range(self.num_agents):
            home_id = random.choice(list(self.places.keys()))
            work_id = random.choice(list(self.places.keys()))
            agent = CityAgent(i, self, home_id, work_id)
            self.schedule.add(agent)
            self.grid.place_agent(agent, self.place_positions[home_id])
            self.places[home_id].add_agent(agent)

        # Infect the first agent
        first_agent = self.schedule.agents[0]
        first_agent.health_status = "Infected"

    def step(self):
        self.schedule.step()

        # Simulate infections in all places
        for place in self.places.values():
            place.simulate_infection(self.infection_probability, self.mask_effectiveness)

        # Toggle time of day
        if self.time_of_day == "morning":
            self.time_of_day = "evening"
        else:
            self.time_of_day = "morning"
```

```python
N = 100  # 50x50 grid
num_agents = 10000
infection_probability = 60  # 60%
latency_period = 3  # 3 days of latency period
infection_duration = 14  # 14 days of infection duration
recovery_rate = 0.95  # 95% recovery rate
mask_effectiveness = 0.7  # 50% effectiveness of masks

model = CityModel(N, num_agents, infection_probability, latency_period, infection_duration, recovery_rate, mask_effectiveness)

for i in range(50):  # Simulate 50 steps
    model.step()
    print(f"Step {i+1}")
    susceptible = sum(agent.health_status == "Susceptible" for agent in model.schedule.agents)
    exposed = sum(agent.health_status == "Exposed" for agent in model.schedule.agents)
    infected = sum(agent.health_status == "Infected" for agent in model.schedule.agents)
    recovered = sum(agent.health_status == "Recovered" for agent in model.schedule.agents)
    deceased = sum(agent.health_status == "Deceased" for agent in model.schedule.agents)
    print(f"Susceptible: {susceptible}, Exposed: {exposed}, Infected: {infected}, Recovered: {recovered}, Deceased: {deceased}")

# 可视化最终的感染状态
def plot_infection(model):
    grid = np.zeros((N, N))
    for agent in model.schedule.agents:
        x, y = agent.pos
        if agent.health_status == "Infected":
            grid[x][y] = 1
        elif agent.health_status == "Exposed":
            grid[x][y] = 0.75
        elif agent.health_status == "Recovered":
            grid[x][y] = 0.5
        elif agent.health_status == "Deceased":
            grid[x][y] = 0.25
    plt.figure(figsize=(10, 10))
    plt.imshow(grid, cmap="viridis", interpolation="nearest")
    plt.title("Infection Status (Yellow = Infected, Light Green = Exposed, Dark Green = Recovered, Brown = Deceased)")
    plt.colorbar()
    plt.show()

plot_infection(model)
```
