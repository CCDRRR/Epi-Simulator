```python
import networkx as nx
import pickle
import matplotlib.pyplot as plt
```

```python
# 导入 pickle 文件
with open("traffic_network.pkl", "rb") as f:
    G_loaded = pickle.load(f)

# 验证导入的网络
print(f"导入的网络节点数: {len(G_loaded.nodes())}")
print(f"导入的网络边数: {len(G_loaded.edges())}")
```

```python
# 绘制网络
plt.figure(figsize=(10, 10))
pos = {node: (data['x'], data['y']) for node, data in G_loaded.nodes(data=True)}
nx.draw(G_loaded, pos, node_size=10, node_color='blue', edge_color='gray', with_labels=False)
plt.show()
```

```python
# 检查节点数据
for node, data in G_loaded.nodes(data=True):
    print(f"Node {node}: {data}")
    break  # 只打印第一个节点的数据来确认
```
